```python
import pandas as pd

df = pd.read_csv('wave21++.txt', delimiter='\t')
df = df.rename(columns={'wave20':'wavelength', 'wave21': 'trace'})
df = df.set_index('wavelength')
df['trace'] += 34.7
```

```python
import matplotlib.pyplot as plt

plt.style.use(['default', './gaeta_light.mplstyle']) # inherit from default to wipe dark background

plt.figure(figsize=(10,4))
plt.plot(df, linewidth=1, label='Swarnav')
plt.ylim(-25)
plt.xlim(1440,1560)
plt.title('Swarnav Comb')
plt.ylabel('Power (dB)')
plt.xlabel('Wavelength')
plt.legend()
plt.show()
```
